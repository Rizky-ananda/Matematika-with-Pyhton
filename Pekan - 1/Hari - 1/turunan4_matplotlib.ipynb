```python
# Library yang digunakan 
import sympy as sy # untuk penyelesaian secara simbolik
import numpy as np # untuk penyelesaian secara numerik
import matplotlib.pyplot as plt # untuk visualiasi data
from scipy.misc import derivative  # Untuk turunan numerik
from scipy.integrate import odeint  # Untuk solusi persamaan diferensial
```

```python
# definisikan variabel
x_value = np.linspace(-5,5,20)
y_value = x_value**3 - 2*x_value**2 + x_value - 5
dy_value = 3*x_value**2 - 4*x_value + 1
x_value
```

```python
plt.plot(x_value, y_value, label="f(x)", color='blue')
plt.plot(x_value, dy_value, label="f'(x)", color='red', linestyle='dashed')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Grafik f(x) dan Turunannya")
plt.show()
```

```python
# definisikan variabel
x = sy.symbols("x")

# definisikan fungsi, f(x) = x^4 - 3x^2 + 2x
func =  x**4 - 3*x**2 + 2*x

# turunan f(x)
d_func = sy.diff(func,x)

# cari akar dari f'(x)
akar = sy.solve(d_func,x)

```

```python
# konversi akar menjadi nilai numerik
numerik_value = [float(val) for val in akar]
numerik_value
```

```python
# definisikan f(x) dan f'(x) dalam bentuk numerik
def f_1(x):
    return x**4 - 3*x**2 + 2*x

def f_2(x):
    return 4*x**3 - 6*x + 2
```

```python
# membuat rentang nilai
x_value = np.linspace(-3,3)

y_value = f_1(x_value)
y_value_2 = f_2(x_value)
```

```python
# Membuat plot
plt.figure(figsize=(10, 6))

# Plot fungsi f(x)
plt.plot(x_value, y_value, label=r'$f(x) = x^4 - 3x^2 + 2x$', color='blue')

# Plot turunan f'(x)
plt.plot(x_value, y_value_2, label=r"$f'(x) = 4x^3 - 6x + 2$", color='green')

# Tandai titik stasioner
for x_value in numerik_value:
    plt.plot(x_value, f_1(x_value), 'ro')  # Titik stasioner dengan warna merah

# Menambahkan label, title, dan grid
plt.title("Grafik Fungsi dan Turunan", fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.axhline(0, color='black',linewidth=0.5)  # Menambahkan garis horizontal di y = 0
plt.axvline(0, color='black',linewidth=0.5)  # Menambahkan garis vertikal di x = 0
plt.grid(True)

# Menampilkan legend untuk membedakan grafik
plt.legend()

# Menampilkan grafik
plt.show()

# Menampilkan titik stasioner
print("Titik stasioner di x =", numerik_value)
```
