```python
# Masukan library yang digunakan
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
```

```python
# definisikan variabel yang digunakan
x , y = sy.symbols(["x","y"])

# persamaan yang akan diselesaikan
func = x**2*y + 3*x*y**2
func
```

```python
dx = sy.diff(func,x) # turunan x
dy = sy.diff(func,y) # turunan y
```

```python
# menampilkan masing masing turunan
print(f"∂f/∂x = {dx}")
print(f"∂f/∂y = {dy}")
```

```python
# subtitusi masing masin turunan, x=1,y=2
df_dx = dx.subs({x:1, y:2})
df_dy = dy.subs({x:1, y:2})
```

```python
# menampilkan hasil
print(f"hasil value dari dx adalah = {df_dx}")
print(f"hasil value dari dy adalah = {df_dy}")
```

```python
# pengerjaan secara numerik
def f(x, y):
    return x**2 * y + 3 * x * y**2

# Titik yang akan dicari
x, y = 1.0, 2.0
dx = 1e-5
dy = 1e-5

# Turunan parsial numerik
df_dx = (f(x + dx, y) - f(x, y)) / dx
df_dy = (f(x, y + dy) - f(x, y)) / dy

```

```python
print("∂f/∂x (numerik) =", df_dx)
print("∂f/∂y (numerik) =", df_dy)
```

```python
# Grid data
x_vals = np.linspace(-2, 2, 50)
y_vals = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

x_vals = np.linspace(-2, 2, 100)
y_vals = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

plt.figure(figsize=(8, 6))
cp = plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(cp)
plt.title('Contour plot of f(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Plot permukaan
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.set_title('Permukaan f(x, y)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```
