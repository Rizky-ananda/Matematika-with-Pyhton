```python
# Sebuah plat logam berbentuk datar dipanaskan secara tidak merata. 
# Setelah beberapa saat, suhu pada setiap titik di permukaan plat
# model fungsi: 50 - x^2 - y^2\
# tentukan suhu pada titik ini, jika x=2,dan y=1
```

```python
# Masukan library yang digunakan
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
```

```python
# definisikan variabel yang digunakan
x,y = sy.symbols(["x","y"])

# persamaan yang akan diselesaikan
func = 50 - x**2 - y**2
func 
```

```python
# turunan pertama x
dx = sy.diff(func,x)
dx
```

```python
# turunan pertama y
dy = sy.diff(func,y)
dy
```

```python
# subtitusi masing masing turunan
df_dx = dx.subs({x:2,y:1})
df_dy = dy.subs({x:2,y:1})
```

```python
# nemampilkan hasil
print(f"value dx adalah: {df_dx}")
print(f"value dy adalah: {df_dy}")
```

```python
# suhu pada titik tersebut
hasil = 50 - df_dx**2 - df_dy**2
hasil
```

```python
# sampai situ juga sudah cukup, karna kebawahnya itu visualisasi dari persamaan tersebut
```

```python
grad_x = dx.subs({x:-4,y:-2})
grad_y = dy.subs({x:-4,y:-2})

gradien_vector = [grad_x,grad_y]
gradien_vector
```

```python
# Fungsi suhu
def T_func(x, y):
    return 50 - x**2 - y**2

# Grid
x_vals = np.linspace(-6, 6, 50)
y_vals = np.linspace(-6, 6, 50)
X, Y = np.meshgrid(x_vals, y_vals)
Z = T_func(X, Y)

# Titik kita dan vektor gradien
x0, y0 = -4, -2
dx, dy = 8, 4  # hasil dari gradien simbolik tadi

# Visualisasi permukaan 3D
fig = plt.figure(figsize=(12, 6))

# Plot permukaan
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax1.scatter(x0, y0, T_func(x0, y0), color='red', s=50, label='Titik (-4, -2)')
ax1.set_title('Permukaan Suhu T(x, y)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('T')
ax1.legend()

# Plot arah gradien (2D)
ax2 = fig.add_subplot(122)
cs = ax2.contourf(X, Y, Z, cmap='viridis')
ax2.quiver(x0, y0, dx, dy, color='red', scale=20)
ax2.set_title('Kontur dan Arah Gradien')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.scatter(x0, y0, color='white', label='Titik (-4, -2)')
ax2.legend()
plt.colorbar(cs, ax=ax2)

plt.tight_layout()
plt.show()
```
