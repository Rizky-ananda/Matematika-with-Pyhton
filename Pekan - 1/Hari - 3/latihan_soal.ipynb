```python
# Masukan Library yang digunakan
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
```

```python
# diberikan sebuah persamaan, 3x^3y^3 - xy^4 + x^2
# dalam persamaan tersebut terdapat 2 variabel x dan y
# oleh karena itu harus turunkan secara parsial
# tentukan turunan ke 1 dari variabel x dan y
# tentukan value dari masing masing turunan jika x=1,y=2
# selesaikan simbolik dan numerik
```

```python
# definisikan variabel yang digunakan
x,y = sy.symbols(["x","y"])

# fungsi yang akan dikerjakan
func = 3*x**3*y**2 - x*y**4 + x**2 
func
```

```python
# turunan pertama x
dx = sy.diff(func,x)
dx
```

```python
# turunan pertam y
dy = sy.diff(func,y)
dy
```

```python
# subtitusi masing masing turunan dengan x=1,y=2
df_dx = dx.subs({x:1,y:2})
df_dy = dy.subs({x:1,y:2})
```

```python
# menampilkan hasil
print(f"value dx adalah: {df_dx}")
print(f"value dy adalah: {df_dy}")
```

```python
# pernyelesaian secara numerik
def f(x,y):
    return 3*x**3*y**2 - x*y**4 +x**2

# titik yang ditentukan
x,y = 1.0,2.0
dx = 1e-5
dy = 1e-5

# menghitung hasil
df_dx = (f(x + dx, y) - f(x, y)) / dx
df_dy = (f(x, y + dy) - f(x, y)) / dy
```

```python
# menampilkan hasil secara numerik
print(f"value dx adalah: {df_dx}")
print(f"value dy adalah: {df_dy}")
```

```python
# visualisasi
# Grid data
x_vals = np.linspace(-2, 2, 50)
y_vals = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

x_vals = np.linspace(-2, 2, 100)
y_vals = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

plt.figure(figsize=(8, 6))
cp = plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(cp)
plt.title('Contour plot of f(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Plot permukaan
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.set_title('Permukaan f(x, y)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```
