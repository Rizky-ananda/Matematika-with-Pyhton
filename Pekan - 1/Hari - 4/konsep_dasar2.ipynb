```python
# sebuah plat logam panas didinginkan dan suhu pada titik tertentu dari plat
# tergantung pada waktu, dan dinyatakan dalam sebuah fungsi:
# T(t) = 100 - 20e^-3x
# cari:
# hitung laju perubahan suhu, pada t=0,5,10
# arti nilai turunan pada wkatu tersebut
# buat grafik
# tentukan waktu saat laju perubahan kurang dari 1oC/s
# setelah itu coba cari jika perubahan suhu < dari 2 derajat/s
```

```python
# Masukan Library yang digunakan
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
```

```python
# definisikan variabel yang digunakan
t = sy.symbols("t")
e = sy.exp

# persamaan yang digunakan
persamaan = 100 - 20*e(-0.3*t)
persamaan # bentuk persamaan yang digunakan ada eksponensial
```

```python
# mencari laju perubahan suhu, t=0,5,10
# laju perubahan suhu dapat dicari dengan menurunkan persamaan yang sudah ada
laju_perubahan_suhu = sy.diff(persamaan,t)
laju_perubahan_suhu # persamaan yang akan digunakan dalam perhitungan
```

```python
# subtitusi persamaan turunan tadi dengan 0,5,10
t0 = laju_perubahan_suhu.subs(t,0)
t5 = laju_perubahan_suhu.subs(t,5)
t10 = laju_perubahan_suhu.subs(t,10)
```

```python
# menampilkan hasil
print(f"laju perubahan suhu pada t0 adalah: {t0:.4f}")
print(f"laju perubahan suhu pada t5 adalah: {t5:.4f}")
print(f"laju perubahan suhu pada t10 adalah: {t10:.4f}")
```

```python
# menentukan waktu saat laju perubahan suhu kurang dari 1Â°C/detik.
# masih menggunakan persamaan yang sama seperti sebelumnya yaitu turunan persamaan yang diketahui
# sehingga persamaannya menjadi 6e^-3t < suhu yang dicari
# untuk menyelesaikannya gunakan ln sehinggan menjadi -ln(suhu/6) / -0.3

suhu_1,suhu_2 = 1,2

detik_perubahan1 = sy.log(suhu_1/6) / -0.3
detik_perubahan2 = sy.log(suhu_2/6) / -0.3
```

```python
print(f"perubahan suhu < dari 1oC terjadi pada detik ke - {detik_perubahan1:.2f}")
print(f"perubahan suhu < dari 2oC terjadi pada detik ke - {detik_perubahan2:.2f}")
```

```python
# mencari suhu di waktu ke n
# persamaan yang digunakan adalah Tlingkungan + (t0 - Tlingkungan) *e^kt
# asumsi tlingkungan = 25
Tlingkungan = 25
Tawal = 100
kons = 0.1
t = 5
suhu_sekarang = Tlingkungan + (Tawal - Tlingkungan) * e(-kons*t)
suhu_sekarang # suhu pada detik ke 5
```

```python
# visualisasi melalui grafik
def temperature(t):
    return Tlingkungan + (Tawal - Tlingkungan) * np.exp(-kons*t)

time = np.linspace(0,100,300)

# hitung suhu
T = temperature(time)

# titik khusus : 5
t_khusus = 15
T_khusus = temperature(t_khusus)

# Plot grafik suhu terhadap waktu
plt.figure(figsize=(10, 6))
plt.plot(time, T, label=f'k = {kons}', color='green')
plt.scatter(t_khusus, T_khusus, color='red', zorder=5, label=f'Titik di t=5s\nSuhu={T_khusus:.2f}Â°C')

# Tambahkan teks keterangan
plt.text(t_khusus, T_khusus - 5, f"t = {t_khusus}s\nT = {T_khusus:.2f}Â°C", 
         ha='center', fontsize=10, color='red', bbox=dict(facecolor='white', edgecolor='red', boxstyle='round'))
plt.title('Pendinginan Suhu (k = 0.1)')
plt.xlabel('Waktu (detik)')
plt.ylabel('Suhu (Â°C)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
