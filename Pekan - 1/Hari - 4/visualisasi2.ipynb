```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi suhu T(t)
def T(t):
    return 100 - 20 * np.exp(-0.3 * t)

# Turunan suhu T'(t)
def dT_dt(t):
    return 6 * np.exp(-0.3 * t)

# Buat array waktu dari 0 s.d 15 detik
t_vals = np.linspace(0, 15, 300)
T_vals = T(t_vals)
dT_vals = dT_dt(t_vals)

# Titik saat dT/dt < 2
threshold = 2
t_below_threshold = t_vals[dT_vals < threshold][0]

# Plot suhu dan laju perubahan
plt.figure(figsize=(12, 6))

# Grafik suhu
plt.subplot(1, 2, 1)
plt.plot(t_vals, T_vals, label="Suhu T(t)", color='blue')
plt.title("Suhu vs Waktu")
plt.xlabel("Waktu (detik)")
plt.ylabel("Suhu (C)")
plt.grid(True)
plt.legend()

# Grafik laju perubahan suhu
plt.subplot(1, 2, 2)
plt.plot(t_vals, dT_vals, label="dT/dt", color='orange')
plt.axhline(y=threshold, color='red', linestyle='--', label="Threshold 2C/s")
plt.axvline(x=t_below_threshold, color='green', linestyle='--', label=f"t = {t_below_threshold:.2f}s")
plt.title("Laju Perubahan Suhu")
plt.xlabel("Waktu (detik)")
plt.ylabel("dT/dt (C/s)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

print(f" Perubahan suhu < 2C/s mulai terjadi setelah t = {t_below_threshold:.2f} detik")
```
