```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Fungsi posisi mobil terhadap waktu
def s(t):
    return 4*t**3 - 3*t**2 + 2*t

# Fungsi kecepatan (turunan pertama dari s)
def v(t):
    return 12*t**2 - 6*t + 2

# Fungsi percepatan (turunan kedua dari s)
def a(t):
    return 24*t - 6
```

```python
# Waktu dalam detik
t1 = 2
t2 = 4
```

```python
# Hitung posisi, kecepatan, dan percepatan
pos_t1 = s(t1)
pos_t2 = s(t2)
jarak_tempuh = pos_t2 - pos_t1
kecepatan = v(t2)
percepatan = a(t2)
```

```python
# Cetak hasil
print(f"Posisi pada t={t1} detik: {pos_t1} meter")
print(f"Posisi pada t={t2} detik: {pos_t2} meter")
print(f"Jarak tempuh antara t={t1} dan t={t2}: {jarak_tempuh} meter")
print(f"Kecepatan mobil pada t={t2} detik: {kecepatan} m/s")
print(f"Percepatan mobil pada t={t2} detik: {percepatan} m/s^2")
```

```python
# Visualisasi posisi terhadap waktu
t = np.linspace(0, 5, 100)
s_values = s(t)

plt.figure(figsize=(8, 5))
plt.plot(t, s_values, label='Posisi s(t)', color='blue')
plt.scatter([t1, t2], [pos_t1, pos_t2], color='red', zorder=5)
plt.title("Grafik Posisi terhadap Waktu")
plt.xlabel("Waktu (detik)")
plt.ylabel("Posisi (meter)")
plt.grid(True)
plt.legend()
plt.show()
```
